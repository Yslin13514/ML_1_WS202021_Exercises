### 1.1 Compare the ML estimates of sample size 50 and 5000.

The Python codes below generate 1) 50 samples of 2 feature vectors and 2) 5000 samples then use ML to estimate the mean (μ_50 and μ_5000) and covariance (c_50 and μ_5000). The ML estimates different sample sizes are listed below 

```python
### 1.1. Generate 50/5000 samples of bivariate Gaussian distribution and estimate 
# the mean and variance using Maximum likelihood method.

sampleN1 = 50
sampleN2 = 5000

meanVec = np.array([2, -2])
covVec = np.array([[0.9, 0.2], [0.2, 0.3]])

# We need to transpose the output so it fits 2 arrays.
rng = np.random.default_rng(seed = 45)
x1, y1 = rng.multivariate_normal(meanVec, covVec, sampleN1).T
x2, y2 = rng.multivariate_normal(meanVec, covVec, sampleN2).T
rxy1 = np.array([x1, y1]).T
rxy2 = np.array([x2, y2]).T

# ML estimates
mu1 = np.mean(rxy1, axis = 0)
mu2 = np.mean(rxy2, axis = 0)
cov1 = (rxy1-mu1).T.dot(rxy1-mu1)/sampleN1
cov2 = (rxy2-mu2).T.dot(rxy2-mu2)/sampleN2

print("mu1 = ", mu1, "mu2 =", mu2, sep = ' ', end = '\n\n')
print("cov1 = ", cov1, "cov2 =", cov2, sep = ' \n')
```

Compared to the original Gaussian distribution,
the ML estimates using 5000 samples are closer to the original distribution values, suggesting that the larger the sample size, the better the estimation results.


### 1.2 3 classes 1000 samples training and testing. 
Next, I sample 1000 data points from three distributions with different mean and covariance values for 1) training and 2) testing data sets (please refer to the exercise sheet for population parameters). We test whether the ML estimates are different with different covariance matrices.

```python
### 1.2. A small classification task: with 3 classes (k = 3). Training and test
# sets both have 1000 samples. Each class has 1/3 samples (so maybe 333, 333, 334)

sampleC1N = round(1000/3);
sampleC2N = round(1000/3);
sampleC3N = 1000 - sampleC1N - sampleC2N

meanVecC1 = np.array([0, 0, 0]).T
meanVecC2 = np.array([1, 2, 2]).T
meanVecC3 = np.array([3, 3, 4]).T

diagCovVec = np.array([[0.8, 0, 0],
                       [0, 0.8, 0],
                       [0, 0, 0.8]]);

nonDiagCovVec = np.array([[0.8, 0.2, 0.1],
                       [0.2, 0.8, 0.2],
                       [0.1, 0.2, 0.8]]);
```

#### 1.2.a ML estimates using diagonal covariance matrix:
I create two sample sets (training and testing), use training data to estimate parameters, then classify testing data based on the Euclidean distances between data points to the estimated means.

```python
## diagonal cov matrix
rng = np.random.default_rng(seed = 45)
# Create samples

xC1tr, yC1tr, zC1tr = rng.multivariate_normal(meanVecC1, diagCovVec, sampleC1N).T
xC2tr, yC2tr, zC2tr = rng.multivariate_normal(meanVecC2, diagCovVec, sampleC2N).T
xC3tr, yC3tr, zC3tr = rng.multivariate_normal(meanVecC3, diagCovVec, sampleC3N).T

xC1te, yC1te, zC1te = rng.multivariate_normal(meanVecC1, diagCovVec, sampleC1N).T
xC2te, yC2te, zC2te = rng.multivariate_normal(meanVecC2, diagCovVec, sampleC2N).T
xC3te, yC3te, zC3te = rng.multivariate_normal(meanVecC3, diagCovVec, sampleC3N).T


rxyzC1tr = np.array([xC1tr, yC1tr, zC1tr]).T;
rxyzC2tr = np.array([xC2tr, yC2tr, zC2tr]).T;
rxyzC3tr = np.array([xC3tr, yC3tr, zC3tr]).T;

rxyzC1te = np.array([xC1te, yC1te, zC1te]).T;
rxyzC2te = np.array([xC2te, yC2te, zC2te]).T;
rxyzC3te = np.array([xC3te, yC3te, zC3te]).T;

muC1tr = np.mean(rxyzC1tr, axis = 0)
muC2tr = np.mean(rxyzC2tr, axis = 0)
muC3tr = np.mean(rxyzC3tr, axis = 0)

covC1tr = (rxyzC1tr-muC1tr).T.dot(rxyzC1tr-muC1tr)/sampleC1N
covC2tr = (rxyzC2tr-muC2tr).T.dot(rxyzC2tr-muC2tr)/sampleC2N
covC3tr = (rxyzC3tr-muC3tr).T.dot(rxyzC3tr-muC3tr)/sampleC3N

comCovtr = np.mean([covC1tr, covC2tr, covC3tr], axis = 0);

print("mu Class1 (diagnal)=", muC1tr,"mu Class2 (diagnal)=", muC2tr,
     "mu Class3 (diagnal)=", muC3tr, sep = ' \n', end = '\n\n')
print("Common covariance matrix (diagnal)=\n", comCovtr, sep=' ')
```

Classification process and results (diagonal):

```python
# Classification: assigning each data point to one class

labelsC1 = np.ones([sampleC1N,1])*1
labelsC2 = np.ones([sampleC2N,1])*2
labelsC3 = np.ones([sampleC3N,1])*3

testAnsC1 = np.zeros([sampleC1N,1])
testAnsC2 = np.zeros([sampleC2N,1])
testAnsC3 = np.zeros([sampleC3N,1])

corrAnsC1 = np.zeros([sampleC1N,1])
corrAnsC2 = np.zeros([sampleC2N,1])
corrAnsC3 = np.zeros([sampleC3N,1])

## Stupid way of doing stuff... but it works, so (shrug)
i = 0
for xyz1 in rxyzC1tr:
    
    #print(xyz1)
    
    distC1 = np.sqrt(np.sum((xyz1-muC1tr)**2))
    distC2 = np.sqrt(np.sum((xyz1-muC2tr)**2))
    distC3 = np.sqrt(np.sum((xyz1-muC3tr)**2))
    
    minDist = np.argmin([distC1, distC2, distC3])
    
    if minDist == 0:
        testAnsC1[i] = 1
        corrAnsC1[i] = 1
    elif minDist == 1:
        testAnsC1[i] = 2
    else:
        testAnsC1[i] = 3
    i = i+1

j = 0
for xyz2 in rxyzC2tr:
    
    #print(xyz1)
    
    distC1 = np.sqrt(np.sum((xyz2-muC1tr)**2))
    distC2 = np.sqrt(np.sum((xyz2-muC2tr)**2))
    distC3 = np.sqrt(np.sum((xyz2-muC3tr)**2))
    
    minDist = np.argmin([distC1, distC2, distC3])
    
    if minDist == 0:
        testAnsC2[j] = 1
    elif minDist == 1:
        testAnsC2[j] = 2
        corrAnsC2[j] = 1
    else:
        testAnsC2[j] = 3
    j = j+1

k = 0
for xyz3 in rxyzC3tr:
    
    #print(xyz1)
    
    distC1 = np.sqrt(np.sum((xyz3-muC1tr)**2))
    distC2 = np.sqrt(np.sum((xyz3-muC2tr)**2))
    distC3 = np.sqrt(np.sum((xyz3-muC3tr)**2))
    
    minDist = np.argmin([distC1, distC2, distC3])
    
    if minDist == 0:
        testAnsC3[k] = 1
    elif minDist == 1:
        testAnsC3[k] = 2
    else:
        testAnsC3[k] = 3
        corrAnsC3[k] = 1
    k = k+1

diagACC = sum([sum(corrAnsC1), sum(corrAnsC2), sum(corrAnsC3)])/1000
print("Classification ACC (diagnal)=", diagACC, sep = ' ')
```

#### 1.2.b ML estimates using non-diagonal covariance matrix:
Doing the same thing again but with non-diagonal covariance matrix.

```python
## non-diagonal cov matrix
rng = np.random.default_rng(seed = 45)

xC1trnd, yC1trnd, zC1trnd = rng.multivariate_normal(meanVecC1, nonDiagCovVec, sampleC1N).T
xC2trnd, yC2trnd, zC2trnd = rng.multivariate_normal(meanVecC2, nonDiagCovVec, sampleC2N).T
xC3trnd, yC3trnd, zC3trnd = rng.multivariate_normal(meanVecC3, nonDiagCovVec, sampleC3N).T

xC1tend, yC1tend, zC1tend = rng.multivariate_normal(meanVecC1, nonDiagCovVec, sampleC1N).T
xC2tend, yC2tend, zC2tend = rng.multivariate_normal(meanVecC2, nonDiagCovVec, sampleC2N).T
xC3tend, yC3tend, zC3tend = rng.multivariate_normal(meanVecC3, nonDiagCovVec, sampleC3N).T


rxyzC1trnd = np.array([xC1trnd, yC1trnd, zC1trnd]).T;
rxyzC2trnd = np.array([xC2trnd, yC2trnd, zC2trnd]).T;
rxyzC3trnd = np.array([xC3trnd, yC3trnd, zC3trnd]).T;

rxyzC1tend = np.array([xC1tend, yC1tend, zC1tend]).T;
rxyzC2tend = np.array([xC2tend, yC2tend, zC2tend]).T;
rxyzC3tend = np.array([xC3tend, yC3tend, zC3tend]).T;

muC1trnd = np.mean(rxyzC1trnd, axis = 0)
muC2trnd = np.mean(rxyzC2trnd, axis = 0)
muC3trnd = np.mean(rxyzC3trnd, axis = 0)

covC1trnd = (rxyzC1trnd-muC1trnd).T.dot(rxyzC1trnd-muC1trnd)/sampleC1N
covC2trnd = (rxyzC2trnd-muC2trnd).T.dot(rxyzC2trnd-muC2trnd)/sampleC2N
covC3trnd = (rxyzC3trnd-muC3trnd).T.dot(rxyzC3trnd-muC3trnd)/sampleC3N

nondiagcomCovtr = np.mean([covC1trnd, covC2trnd, covC3trnd], axis = 0);
print("mu Class1 (nondiagnal)=", muC1trnd,"mu Class2 (nondiagnal)=", muC2trnd,
     "mu Class3 (nondiagnal)=", muC3trnd, sep = ' \n', end = '\n\n')
print("Common covariance matrix (nondiagnal)=\n", nondiagcomCovtr, sep=' ', end='\n\n')

# Classification: assigning each data point to one class

labelsC1nd = np.ones([sampleC1N,1])*1
labelsC2nd = np.ones([sampleC2N,1])*2
labelsC3nd = np.ones([sampleC3N,1])*3

testAnsC1nd = np.zeros([sampleC1N,1])
testAnsC2nd = np.zeros([sampleC2N,1])
testAnsC3nd = np.zeros([sampleC3N,1])

corrAnsC1nd = np.zeros([sampleC1N,1])
corrAnsC2nd = np.zeros([sampleC2N,1])
corrAnsC3nd = np.zeros([sampleC3N,1])

## Stupid way of doing stuff... but it works, so (shrug)
i = 0
for xyz1 in rxyzC1trnd:
    
    #print(xyz1)
    
    distC1 = np.sqrt(np.sum((xyz1-muC1trnd)**2))
    distC2 = np.sqrt(np.sum((xyz1-muC2trnd)**2))
    distC3 = np.sqrt(np.sum((xyz1-muC3trnd)**2))
    
    minDist = np.argmin([distC1, distC2, distC3])
    
    if minDist == 0:
        testAnsC1nd[i] = 1
        corrAnsC1nd[i] = 1
    elif minDist == 1:
        testAnsC1nd[i] = 2
    else:
        testAnsC1nd[i] = 3
    i = i+1

j = 0
for xyz2 in rxyzC2trnd:
    
    #print(xyz1)
    
    distC1 = np.sqrt(np.sum((xyz2-muC1trnd)**2))
    distC2 = np.sqrt(np.sum((xyz2-muC2trnd)**2))
    distC3 = np.sqrt(np.sum((xyz2-muC3trnd)**2))
    
    minDist = np.argmin([distC1, distC2, distC3])
    
    if minDist == 0:
        testAnsC2nd[j] = 1
    elif minDist == 1:
        testAnsC2nd[j] = 2
        corrAnsC2nd[j] = 1
    else:
        testAnsC2nd[j] = 3
    j = j+1

k = 0
for xyz3 in rxyzC3trnd:
    
    #print(xyz1)
    
    distC1 = np.sqrt(np.sum((xyz3-muC1trnd)**2))
    distC2 = np.sqrt(np.sum((xyz3-muC2trnd)**2))
    distC3 = np.sqrt(np.sum((xyz3-muC3trnd)**2))
    
    minDist = np.argmin([distC1, distC2, distC3])
    
    if minDist == 0:
        testAnsC3nd[k] = 1
    elif minDist == 1:
        testAnsC3nd[k] = 2
    else:
        testAnsC3nd[k] = 3
        corrAnsC3nd[k] = 1
    k = k+1

nondiagACC = sum([sum(corrAnsC1nd), sum(corrAnsC2nd), sum(corrAnsC3nd)])/1000
print("Classification ACC (non diagnal)=", nondiagACC, sep = ' ')
```

It can be seen that the classification accuracy is lower in the case of non-diagonal covariance matrix. This suggests the impact of correlation between different classes. If classes are correlated, it might be harder for the classifier to discriminate between the data from different classes.

## 2. Expectation Maximization

In the second part, I sample data from a 2-dimensional mixed Gaussian distribution (with three components), and do EM estimates for the unknown (pretended that we don't know the original parameters) parameters either with no initial values or different initial values.

```python
import numpy as np
from scipy.stats import multivariate_normal as m_normal
from sklearn.mixture import GaussianMixture
```

First, I create sample data from three different mutivariate Gaussian distributions. The sample number from each distribution depends on the predetermined weights (P).

```python
n_samples = 600
n_dim = 2

P = [0.4, 0.4, 0.2]
I = np.array([[1, 0], [0,1]])
C1 = 0.1*I
C2 = 0.2*I
C3 = 0.3*I
Mu1 = [1,1]
Mu2 = [3,3]
Mu3 = [2,6]

mn1 = m_normal(mean = Mu1, cov = C1)
mn2 = m_normal(mean = Mu2, cov = C2)
mn3 = m_normal(mean = Mu3, cov = C3)

# Take 0.4*600 from the first distribution, 0.4*600 from the second and so on.
mn1Samples2 = mn1.rvs(size = round(n_samples*P[0]), random_state = 12345)
mn2Samples2 = mn1.rvs(size = round(n_samples*P[1]), random_state = 12345)
mn3Samples2 = mn1.rvs(size = round(n_samples*P[2]), random_state = 12345)

totalSamples = np.r_[mn1Samples2, mn2Samples2, mn3Samples2]
```

### 2.1 Use EM to estimate the parameters without providing initial values.
https://scikit-learn.org/stable/modules/generated/sklearn.mixture.GaussianMixture.html

I choose 'random' option instead of 'kmeans'.


```python
## Fit without initial parameters

fit1 = GaussianMixture(n_components = 3, covariance_type='full'\
                       ,init_params = 'random', random_state=12345).fit(totalSamples)
print("Means (3 components)=\n",fit1.means_, sep = ' ', end = '\n\n')
print("Covariances =\n", fit1.covariances_, sep = ' ', end = '\n\n')
print("Weights = \n", fit1.weights_, end='\n\n')
```

The estimated means and covariances of all components are very similar to each other. Interestingly, if I use init_params = 'kmeans' instead, the results differ quite a lot, especially the covariances and weights. The weights are more similar to the original values.

```python
fit1 = GaussianMixture(n_components = 3, covariance_type='full'\
                       ,init_params = 'kmeans', random_state=12345).fit(totalSamples)
print("Means (3 components)=\n",fit1.means_, sep = ' ', end = '\n\n')
print("Covariances =\n", fit1.covariances_, sep = ' ', end = '\n\n')
print("Weights = \n", fit1.weights_, end='\n\n')
```

### 2.2 Use EM to estimate the parameters with initial values.

```python
## Fit with initial parameters

wInit1 = np.array([1/3, 1/3, 1/3]).T;
muInit1 = np.array([[0, 2], [5,2], [5, 5]]);
preInit1 =np.array([np.linalg.inv(0.15*I), np.linalg.inv(0.27*I),\
                    np.linalg.inv(0.4*I)])

fit2 = GaussianMixture(n_components = 3, covariance_type='full'\
                       ,init_params = 'random', random_state=12345, weights_init = wInit1\
    , means_init = muInit1, precisions_init = preInit1)\
    .fit(totalSamples)
print("Means (3 components)=\n",fit2.means_, end = '\n\n')
print("Covariances =\n",fit2.covariances_, end = '\n\n')
print("Weights = \n",fit2.weights_, end = '\n\n')
```

Compared to the previous fitting results, it seems that the estimated covariance matrices deviate a lot from the original sample data. The weight estimates are very biased towards the first Gaussian distribution.


### 2.3 Use EM to estimate the parameters with initial values and only two components.

```python
## Fit with 2 components

wInit2 = np.array([1/2, 1/2]).T;
muInit2 = np.array([[1.6, 1.4], [1.4,1.6]]);
preInit2 =np.array([np.linalg.inv(0.2*I), np.linalg.inv(0.4*I)])

fit3 = GaussianMixture(n_components = 2, covariance_type='full'\
                       ,init_params = 'random', random_state=12345, weights_init = wInit2\
    , means_init = muInit2, precisions_init = preInit2)\
    .fit(totalSamples)
print("Means (2 components)=\n",fit3.means_, end = '\n\n')
print("Covariances =\n",fit3.covariances_, end = '\n\n')
print("Weights = \n",fit3.weights_, end = '\n\n')
```

Compared to 2.2 and 2.1, the results of EM estimation with 2 components and different initial values change the final fitting outcomes even more. To sum up, initial values could drastically change the fitting results in the current case.

## 3. Clustering

In the third part, I sample 1000 data points from a population composed of four 2-dimensional mixed Gaussian distributions then use K-means method to estimate the cluster representatives parameters (the locations of the cluster center).

```python
import numpy as np
from scipy.stats import multivariate_normal as m_normal
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.cluster import KMeans
plt.style.use('seaborn')
```

```python

n_samples = 1000
n_clusters = 4
n_eachGroup = round(n_samples/n_clusters)

mu1 = [0, 0];
mu2 = [10, 0];
mu3 = [0, 6];
mu4 = [9, 8];

c1 = np.array([[1, 0],[0, 1]])
c2 = np.array([[1, 0.2], [0.2, 1.5]])
c3 = np.array([[1, 0.4], [0.4, 1.1]])
c4 = np.array([[0.3, 0.2], [0.2, 0.5]])

cG1 = m_normal(mean = mu1, cov = c1)
cG1Samples = cG1.rvs(size = n_eachGroup, random_state = 12345)
cG2 = m_normal(mean = mu2, cov = c2)
cG2Samples = cG2.rvs(size = n_eachGroup, random_state = 12345)
cG3 = m_normal(mean = mu3, cov = c3)
cG3Samples = cG3.rvs(size = n_eachGroup, random_state = 12345)
cG4 = m_normal(mean = mu4, cov = c4)
cG4Samples = cG4.rvs(size = n_eachGroup, random_state = 12345)

totalSamples4C = np.r_[cG1Samples, cG2Samples, cG3Samples, cG4Samples]

plt.scatter(totalSamples4C[:, 0], totalSamples4C[:, 1], s=20);# No color yet
```

Here, by plotting the sampled data we can already see clearly that there are four clusters without doing further analysis. I thus predict that the C=4 k-means fitting should come up with the best results.

### 3.1 C = 4 without providing initial cluster centers.


```python
# 1. C = 4;
nC = 4;
est = KMeans(n_clusters=nC)  # 4 clusters
est.fit(totalSamples4C)
y_kmeans = est.predict(totalSamples4C)
centers = est.cluster_centers_;

fig, ax = plt.subplots()
ax.scatter(totalSamples4C[:, 0], totalSamples4C[:, 1], c=y_kmeans, s=20, cmap='rainbow');
ax.scatter(centers[:, 0], centers[:, 1], c='black', s=50, alpha=0.5);
for i in range(nC):
    txt = str([round(centers[i,0], 2), round(centers[i,1], 2)])
    print(txt)
    ax.annotate(txt, (centers[i, 0]+0.2, centers[i, 1]+0.2))
```

It can be seen that the k-means algorithms do a great job identifying the four clusters and estimate with small errors the cluster representatives.

### 3.2 C = 3 without providing initial cluster centers.

```python
# 2. C = 3;
nC = 3;
est = KMeans(n_clusters=nC)  # 4 clusters
est.fit(totalSamples4C)
y_kmeans = est.predict(totalSamples4C)
centers = est.cluster_centers_;

fig, ax = plt.subplots()
ax.scatter(totalSamples4C[:, 0], totalSamples4C[:, 1], c=y_kmeans, s=20, cmap='rainbow');
ax.scatter(centers[:, 0], centers[:, 1], c='black', s=50, alpha=0.5);
for i in range(nC):
    txt = str([round(centers[i,0], 2), round(centers[i,1], 2)])
    print(txt)
    ax.annotate(txt, (centers[i, 0]+0.2, centers[i, 1]+0.2))
```

It can be seen that when C was set to 3, the K-means algorithms combine the two closest clusters ([0,0], [0,6]) together as one cluster while keeping the two other clusters as they are. This demonstrates the weakness of k-mean clustering method: as we need to predefine how many clusters to find, the results will be biased or fail if the guessed number of cluster deviates too much from the reality. But this could be fixed by comparing the outcome of many clustering results. We can also use validation method such as silhouette testing (https://en.wikipedia.org/wiki/Silhouette_(clustering)) to detect outliers and determine how well each item is classified to each cluster.

### 3.3 C = 5 without providing initial cluster centers.

```python
# 3. C = 5;
nC = 5
est0 = KMeans(n_clusters = nC)  # 5 clusters
est0.fit(totalSamples4C)
y_kmeans0 = est0.predict(totalSamples4C)
centers0 = est0.cluster_centers_;

fig0, ax0 = plt.subplots()
ax0.scatter(totalSamples4C[:, 0], totalSamples4C[:, 1], c=y_kmeans0, s=20, cmap='rainbow');
ax0.scatter(centers0[:, 0], centers0[:, 1], c='black', s=50, alpha=0.5);
for i in range(nC):
    txt = str([round(centers0[i,0], 2), round(centers0[i,1], 2)])
    print(txt)
    ax0.annotate(txt, (centers0[i, 0]+0.2, centers0[i, 1]+0.2))
```

As I predetermine there should be 5 clusters this time, the algorithms reach the goal by dividing the cluster with higher covariance matrix (center at [10,0]) into two groups.

### 3.4 C = 4 with initial cluster parameters.

```python
# 4. C = 4, but with initial values
nC = 4
initialCenters = np.array([[-2,-2],[-2.1, -2.1], [-2, -2.2], [-2.1, -2.2]])
est1 = KMeans(n_clusters=nC, init=initialCenters)  # 4 clusters
est1.fit(totalSamples4C)
y_kmeans1 = est1.predict(totalSamples4C)
centers1 = est1.cluster_centers_;

fig1, ax1 = plt.subplots()
ax1.scatter(totalSamples4C[:, 0], totalSamples4C[:, 1], c=y_kmeans1, s=20, cmap='rainbow');
ax1.scatter(centers1[:, 0], centers1[:, 1], c='black', s=50, alpha=0.5);
for i in range(nC):
    txt = str([round(centers1[i,0], 2), round(centers1[i,1], 2)])
    print(txt)
    ax1.annotate(txt, (centers1[i, 0]+0.2, centers1[i, 1]+0.2))
```

Compared to 3.1, only the order of the clusters (color-coded) is changed as a result. 

### 3.5 C = 4 with different initial cluster parameters.

```python
# 5. C = 4, but with bad initial values
nC = 4
initialCenters = np.array([[-2,-2],[-2.1, -2.1], [-2, -2.2], [30, 30]])
est2 = KMeans(n_clusters=nC, init=initialCenters)  # 4 clusters
est2.fit(totalSamples4C)
y_kmeans2 = est2.predict(totalSamples4C)
centers2 = est2.cluster_centers_;

fig2, ax2 = plt.subplots()
ax2.scatter(totalSamples4C[:, 0], totalSamples4C[:, 1], c=y_kmeans2, s=20, cmap='rainbow');
ax2.scatter(centers2[:, 0], centers2[:, 1], c='black', s=50, alpha=0.5);
for i in range(nC):
    txt = str([round(centers2[i,0], 2), round(centers2[i,1], 2)])
    print(txt)
    ax2.annotate(txt, (centers2[i, 0]+0.2, centers2[i, 1]+0.2))
```

Again, even though the initial position of one cluster is very far from the real centers, the final estimate of the centers are the same as 3.1. This could suggest that k-means is more robust when it comes to initial values.
